```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pickle
```

# Data collection and preprocessing

```python
df = pd.read_csv('./dataset.csv', header=None)

df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df[60].value_counts()
```

```python
df.groupby(60).mean()
```

```python
X = df.drop(columns=60, axis=1)
Y = df[60]
```

```python
print(X)
print(Y)
```

### Training & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.1, stratify=Y, random_state=30)
```

```python
X.shape
```

```python
X_train.shape, Y_train.shape, Y_train[0]
```

Model Training --> Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_preds = model.predict(X_train)
```

```python
training_data_acc = accuracy_score(X_train_preds, Y_train)
print('Accuracy on training data : ', training_data_acc)
```

```python
X_test_preds = model.predict(X_test)
test_data_acc = accuracy_score(X_test_preds, Y_test)

print('Accuracy on testing data : ', test_data_acc)
```
